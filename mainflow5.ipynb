```python
# 1 how many people have heart diseas and how many dont
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind='bar', color=["salmon", "orchid"])
plt.title("Heart diseas values")
plt.xlabel("1= Heart diseas, 0=No heart diseas")
plt.ylabel("Amount")
```

```python
df.target.value_counts().plot(kind='pie', figsize=(8,6))
plt.legend(["Diseas", "No Diseas"]);
```

```python
# check how many men and women 
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind='pie', figsize=(8,6))
plt.legend(["Men", "Women"]);
plt.title("Male female ratio")
```

```python
# 2 people of sex has most heart diseas
pd.crosstab(df.target,df.sex)
```

```python
srn.countplot(x= 'target', data=df, hue='sex')
plt.title("heart diseasnfrequency for male and female")
plt.xlabel("0= No heart diseas, 1=Heart disseas");
```

```python
# 3 people of which sex has which type of chest pain most
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='bar', color=["salmon", "orchid","pink","blue"])
plt.title("chest pain type vs count")
```

```python
pd.crosstab(df.sex,df.cp).plot(kind='bar', color=["salmon", "orchid","pink","blue"])
plt.title("Type of chest pain for sex")
plt.xlabel("0= female, 1=male");
```

```python
# 4 people with which chest pain aer most pron tohave heart diseas
pd.crosstab(df.cp,df.target)
```
